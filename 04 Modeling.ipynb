```python
import pandas as pd
import numpy as np

import pickle

from sklearn.linear_model import LinearRegression, ElasticNet
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import f_classif


import matplotlib.pyplot as plt
import seaborn as sns

from scipy.stats import normaltest

import scipy.stats as stats
import statsmodels.api as sm

```

```python
# X underwent polynomial feature transformation and scaling in feat
X = pickle.load(open('features_to_plot.p','rb'))
y = pickle.load(open('target.p','rb'))
```

```python
lm = LinearRegression()
kf = KFold(n_splits=5, shuffle=True, random_state = 71)
cross_val_score(lm, X, y, cv=kf, scoring='r2')
```

```python
# hold out 20% of the data for final testing
X, X_test, y, y_test = train_test_split(X, y, test_size=.2, random_state=100)
```

```python
lamb = []
alpha = []
lm_reg_mse =[]
lm_reg_r2=[]
for n in np.linspace(0,2,200):
    for k in np.linspace(0,1,10):
        lm_reg =ElasticNet(alpha =n,l1_ratio=k)
        lamb.append(n)
        alpha.append(k)
        lm_reg_mse.append(np.mean(cross_val_score(lm_reg, X, y, cv=kf, scoring='neg_mean_squared_error')))
        lm_reg_r2.append(np.mean(cross_val_score(lm_reg, X, y, cv=kf, scoring='r2')))
```

```python
sns.lineplot(x=lamb, y=lm_reg_mse)
```

```python
sns.lineplot(x=lamb, y=lm_reg_r2);
```

```python
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)
sns.lineplot(x=lamb, y=lm_reg_r2);
plt.title("Model Performance by 5-fold Cross Validation")
plt.xlabel("Lamba Tuning Grid")
plt.ylabel("R Squared");
plt.savefig('Tuning Grid.svg',bbox_inches="tight")
```

```python
elastic_net_results = pd.DataFrame({
    'lamb': lamb,
    'alpha': alpha,
    'lm_reg_mse': lm_reg_mse,
    'lm_reg_r2': lm_reg_r2
}
)
```

```python
elastic_net_results.sort_values(by= 'lm_reg_mse', ascending = False).head(5)
```

```python
elastic_net_final = ElasticNet(random_state=10, alpha =0.110533, l1_ratio =.1111)
elastic_net_final.fit(X,y)
```

```python
pred = elastic_net_final.predict(X_test)
res = y_test - pred
```

```python
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)
plt.scatter(pred, res)
plt.title("Residual plot")
plt.xlabel("Prediction")
plt.ylabel("Residuals");
plt.savefig('Residual Plot.svg',bbox_inches="tight")
```

```python
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)
plt.scatter(pred, y_test)
plt.title("Prediction vs. Actual")
plt.xlabel("Actual Win Percentage")
plt.ylabel("Predicted Win Percentage");
plt.savefig('Prediction.svg',bbox_inches="tight")
```

```python
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)
stats.probplot(res, dist="norm", plot=plt)
plt.title("Normal Q-Q plot");
plt.savefig('Residual Plot QQ.svg',bbox_inches="tight")
```
