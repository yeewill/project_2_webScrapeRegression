```python
import pandas as pd
import pickle

from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import LinearRegression, Ridge, ElasticNet

import scipy.stats as stats
import statsmodels.api as sm

import seaborn as sns
```

```python
# import new era data and creaste dummy variables
new_era = pickle.load(open('ready_for_model_new_era.p','rb'))
new_era['new_era'] = 1
# import old era data and create dummy variables
old_era = pickle.load(open('ready_for_model_old_era.p','rb'))
old_era['new_era'] = 0
```

```python
combined_data = pd.concat([new_era,old_era])
combined_data['win_per'] = combined_data['win_per'] *100
```

```python
combined_data = combined_data.drop('year', axis =1)
```

```python
X , y = combined_data.drop('win_per',axis =1), combined_data['win_per']
```

```python
X['intercept'] = 1
```

```python
sns.pairplot(combined_data);
```

```python
# Create your model
model = sm.OLS(y, X)

# Fit your model to your training set
fit = model.fit()

# Print summary statistics of the model's performance
fit.summary()
```

```python
# Scale X features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X.drop(['intercept','new_era'],axis =1))
```

```python
# Create your model
model = sm.OLS(y, X_scaled)

# Fit your model to your training set
fit = model.fit()

# Print summary statistics of the model's performance
fit.summary()
```
